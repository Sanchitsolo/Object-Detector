```python
from lxml import etree
from PIL import Image
import os
import shutil, sys

import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image
import numpy as np
from numpy import nan
import pandas as pd
```

```python
filename='exp.csv'
maindf=pd.read_csv(filename, sep=',')
num=[]
nm=list(maindf.iloc[:,0])
for k in nm:
    k=k.split('.')[0]
    num.append(k)
```

```python
trainvalb={}
trainvalp={}
trainvalv={}
trainvalw={}
size=int(len(maindf.iloc[:,0])/100*95)
with open('data/VOCdevkit2007/VOC2007/ImageSets/Layout/trainval.txt', 'w+') as f: 
    for i in range(0,size):
        f.write('%s\n' % i)
        trainvalb[i]=maindf.iloc[i,1]
        trainvalp[i]=maindf.iloc[i,2]
        trainvalv[i]=maindf.iloc[i,3]
        trainvalw[i]=maindf.iloc[i,4]
```

```python
testb={}
testp={}
testv={}
testw={}
size=int(len(maindf.iloc[:,0])/100*5)
np.random.randint(0,len(num),size)
with open('data/VOCdevkit2007/VOC2007/ImageSets/Layout/test.txt', 'w+') as f:
    for i in range(size, len(num)):
        f.write('%s\n' % i)
        testb[i]=maindf.iloc[i,1]
        testp[i]=maindf.iloc[i,2]
        testv[i]=maindf.iloc[i,3]
        testw[i]=maindf.iloc[i,4]
```

```python
trainb={}
trainp={}
trainv={}
trainw={}
t_size=int(size/100*90)
with open('data/VOCdevkit2007/VOC2007/ImageSets/Layout/train.txt', 'w+') as f:
    for i in range(0,t_size):
        f.write('%s\n' % i)
        trainb[i]=maindf.iloc[i,1]
        trainp[i]=maindf.iloc[i,2]
        trainv[i]=maindf.iloc[i,3]
        trainw[i]=maindf.iloc[i,4]
```

```python
valb={}
valp={}
valv={}
valw={}
with open('data/VOCdevkit2007/VOC2007/ImageSets/Layout/val.txt', 'w+') as f:
    for i in range(t_size,size):
        f.write('%s\n' % i)
        valb[i]=maindf.iloc[i,1]
        valp[i]=maindf.iloc[i,2]
        valv[i]=maindf.iloc[i,3]
        valw[i]=maindf.iloc[i,4]
```
